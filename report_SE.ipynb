```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid', font_scale=1.3, palette='Set2')
%matplotlib inline
```

```python

# загрузим датасет
time_data = pd.read_csv('experiment.csv')
# выведем первые 5 строк
time_data.head(5)
```

```python
numder_of_tests = 15
data = time_data.groupby(['login', 'experiment'], sort=False).sum() / numder_of_tests
data
```

```python
data = data.reset_index()
```

```python
plt.figure(figsize=(20, 16))
for i in data.groupby('login'):
  plt.scatter(data[data['login'] == i[0]]['experiment'], data[data['login'] == i[0]]['time'], s=15)
  plt.plot(data[data['login'] == i[0]]['experiment'], data[data['login'] == i[0]]['time'], label=i[0])
plt.ylabel('time (ns)')
plt.xlabel('size')
plt.legend(fontsize='small')
```
